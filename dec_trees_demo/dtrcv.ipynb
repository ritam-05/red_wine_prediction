```python
#boston houuse pricing
import numpy as np
import pandas as pd

```

```python
df=pd.read_csv("BostonHousing.csv")
df.head()
```

```python
#description
df.describe()
```

```python
#independent and dependent features
x=df.drop("medv",axis=1)
y=df["medv"]
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#decision tree regressor
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
y_pred=dtr.predict(x_test)
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("MAE:",mae)
print("MSE:",mse)
print("R2 Score:",r2)
```

```python
#hyperparameter tuning using cross validation
from sklearn.model_selection import GridSearchCV
param_grid={'max_depth':[2,4,6,8,10,None],
            'min_samples_split':[2,5,10],
            'min_samples_leaf':[1,2,4],
            'criterion':['mse','friedman_mse','poisson'],
            'splitter':['best','random']}

dtr_cv=DecisionTreeRegressor()
grid_search=GridSearchCV(estimator=dtr_cv,param_grid=param_grid,cv=5,scoring='neg_mean_squared_error',n_jobs=-1,verbose=2)
grid_search.fit(x_train,y_train)    
```

```python
#best parameters
best_params=grid_search.best_params_
print("Best Parameters:",best_params)
```

```python
y_pred_cv=grid_search.predict(x_test)
```

```python
#r2
r2_cv=r2_score(y_test,y_pred_cv)
print("R2 Score after CV:",r2_cv)
```
